1. . What is the difference between a function and a method in Python?
- In Python, functions and methods are both blocks of reusable code, but they differ in how they are defined and used.
>>- Functions:A function is a standalone block of code that is not associated with any particular object.You define it using the def keyword and call it directly by its name.
>>-ex:
def greet(name):
    return f"Hello, {name}!"
print(greet("Alice"))
>>- Methods:A method is a function that is associated with an object and is called on that object. Methods are defined inside a class and operate on the attributes of the object.A method implicitly takes the object (self) as its first parameter when called.
>>-ex:
class Person:
    def greet(self):
        return "Hello!"
person = Person()
print(person.greet())

6. Explain the concept of generators in Python and how they are defined.
- Generators in Python are a special type of iterable that allow you to produce values one at a time, as you need them, rather than generating all values at once and storing them in memory. This makes them highly memory-efficient, especially when dealing with large datasets or infinite sequences.
Generators are defined using functions with the  statement. Unlike regular functions that return a single value using , a generator function can pause its execution and resume later, yielding values one by one.
  -ex:def simple_generator():
    yield 1
    yield 2
    yield 3
gen = simple_generator()
print(next(gen))  # Output: 1
print(next(gen))  # Output: 2
print(next(gen))  # Output: 3

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
- >>- map():Transforms each item in an iterable by applying a specified function.Returns a map object (an iterable) containing the transformed values.
 -ex:numbers = [1, 2, 3, 4]

numbers = [1, 2, 3, 4]
product = reduce(lambda x, y: x * y, numbers)
print(product)

```python
#4.  Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(number):
    if number < 2:
        return False
    for i in range(2, int(number ** 0.5) + 1):
        if number % i == 0:
            return False
    return True
for num in range(1, 201):
    if is_prime(num):
        print(f"{num} is a prime number.")
```

```python
#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
#terms.
class FibonacciIterator:
    def __init__(self, num_terms):
        self.num_terms = num_terms
        self.current_term = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.current_term >= self.num_terms:
            raise StopIteration

        if self.current_term == 0:
            self.current_term += 1
            return self.a
        elif self.current_term == 1:
            self.current_term += 1
            return self.b
        else:
            fib_number = self.a + self.b
            self.a, self.b = self.b, fib_number
            self.current_term += 1
            return fib_number
num_terms = 10
fibonacci_sequence = FibonacciIterator(num_terms)

print("Fibonacci Sequence:")
for number in fibonacci_sequence:
    print(number)
```

```python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def powers_of_two(max_exponent):
    for exponent in range(max_exponent + 1):
        yield 2 ** exponent
max_exponent = 5
for power in powers_of_two(max_exponent):
    print(power)
```

```python
#7. Implement a generator function that reads a file line by line and yields each line as a string.
def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()
```

```python
#9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32
celsius_temperatures = [0, 10, 20, 30, 40]
fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))
print("Temperatures in Fahrenheit:", fahrenheit_temperatures)
```

```python
#10.  Create a Python program that uses `filter()` to remove all the vowels from a given string.
def remove_vowels(input_string):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda char: char not in vowels, input_string))
input_string = "This is an example sentence."
result = remove_vowels(input_string)
print(result)
```

```python
#11. Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this:
# Order Number      Book Title and Author                      Quantity           Price per Item
# 34587             Learning Python,Mark Lutz                     4                    40.95
# 98762             Programming Python,Mark Lutz                  5                    56.80
# 77226             Head First Python,Paul Barry                  3                    32.95
# 88112             Einfuhrung in Python3,Bernd Klein             3                    24.99
#Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
#product of the price per item and the quantity. The product should be increased by 10,-  if the value of the
#order is smaller than 100,00 .
# List of orders with sublists
orders = [
    [34587, "Learning Python,Mark Lutz", 4, 40.95],
    [98762, "Programming Python,Mark Lutz", 5, 56.80],
    [77226, "Head First Python,Paul Barry", 3, 32.95],
    [88112, "Einfuhrung in Python3,Bernd Klein", 3, 24.99],
]
def calculate_total(orders):
    result = []
    for order in orders:
        order_number = order[0]
        quantity = order[2]
        price_per_item = order[3]
        total = quantity * price_per_item
        if total < 100:
            total += 10

        result.append((order_number, round(total, 2)))
    return result
output = calculate_total(orders)
print(output)
```

```python
#11. Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this:
# Order Number      Book Title and Author                      Quantity           Price per Item
# 34587             Learning Python,Mark Lutz                     4                    40.95
# 98762             Programming Python,Mark Lutz                  5                    56.80
# 77226             Head First Python,Paul Barry                  3                    32.95
# 88112             Einfuhrung in Python3,Bernd Klein             3                    24.99

# Write a Python program using lamda and map.
orders = [
    [34587, "Learning Python,Mark Lutz", 4, 40.95],
    [98762, "Programming Python,Mark Lutz", 5, 56.80],
    [77226, "Head First Python,Paul Barry", 3, 32.95],
    [88112, "Einfuhrung in Python3,Bernd Klein", 3, 24.99],
]
result = list(map(lambda order: (order[0],
                                 round(order[2] * order[3] + (10 if order[2] * order[3] < 100 else 0), 2)), orders))
print(result)
```
